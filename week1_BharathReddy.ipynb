### 1.3 Exceptions

Write `safe_divide(a, b)` that:
- Returns `a / b` normally
- Returns `None` if `b == 0` (catch `ZeroDivisionError`)
- Raises `TypeError` with message `"Inputs must be numeric"` if either input is not a number

````python
def safe_divide(a, b):
    """
    Divide a by b with graceful handling of two failure modes:
      - Non-numeric inputs  → raise TypeError
      - Division by zero    → return None instead of crashing
    """
    # Type check
    if not isinstance(a, (int, float)) or not isinstance(b, (int, float)):
        raise TypeError("Inputs must be numeric")

    try:
        return a / b
    except ZeroDivisionError:
        return None

# Tests
assert safe_divide(10, 2) == 5.0
assert safe_divide(5, 0) is None
try:
    safe_divide('x', 2)
    assert False, "Should have raised TypeError"
except TypeError as e:
    assert str(e) == "Inputs must be numeric"
print("1.3 passed ✓")
````

### 4.3 Eigenvalues & Eigenvectors

For the matrix:
```
A = [[4, 1],
     [2, 3]]
```

1. Compute eigenvalues and eigenvectors using `np.linalg.eig`.
2. **Verify** the eigenvector equation: `A @ v = λ * v` for each eigenpair.
3. Plot both eigenvectors as arrows, scaled by their eigenvalues, on a 2D plot.
4. **Explain** in a markdown cell (below): what does it geometrically mean for a matrix to stretch a vector?

````python
A = np.array([[4, 1], [2, 3]], dtype=float)

# Eigenvectors
eigenvalues, eigenvectors = np.linalg.eig(A)

print(f"Eigenvalues: {eigenvalues}")
print(f"Eigenvectors (columns):\n{eigenvectors}")

# Verify equation
print("\nVerification — A @ v  vs  λ × v:")
for i in range(len(eigenvalues)):
    lam = eigenvalues[i]
    vec = eigenvectors[:, i]          # i-th column = i-th eigenvector
    Av  = A @ vec
    lv  = lam * vec
    match = np.allclose(Av, lv)
    print(f"  λ={lam:.2f} | A@v={Av.round(4)}  λv={lv.round(4)}  ✓={match}")

# Plot arrows
colors = ['tomato', 'steelblue']
fig, ax = plt.subplots(figsize=(6, 6))
for i in range(len(eigenvalues)):
    lam = eigenvalues[i]
    vec = eigenvectors[:, i] * lam    # scale by eigenvalue for visual emphasis
    ax.quiver(0, 0, vec[0], vec[1],
              angles='xy', scale_units='xy', scale=1,
              color=colors[i], width=0.015,
              label=f'λ={lam:.1f}  v={eigenvectors[:, i].round(2)}')

ax.set_xlim(-4, 4); ax.set_ylim(-4, 4)
ax.axhline(0, color='k', lw=0.8); ax.axvline(0, color='k', lw=0.8)
ax.set_xlabel('x'); ax.set_ylabel('y')
ax.set_title('Eigenvectors Scaled by Eigenvalues')
ax.legend(); ax.set_aspect('equal')
plt.tight_layout(); plt.show()

print("4.3 passed ✓")
````

**Your Geometric Explanation (4.3):**

> When a matrix **A** is applied to most vectors it both rotates and scales them — the output points in a completely different direction. An **eigenvector** is special: multiplying it by **A** keeps it pointing in exactly the same direction (or exactly opposite). The corresponding **eigenvalue** λ tells you *how much* the matrix stretches (λ > 1), shrinks (0 < λ < 1), flips (λ < 0), or leaves unchanged (λ = 1) that vector. Visually, think of the matrix as a rubber-sheet transformation: eigenvectors are the axes along which the sheet is pulled or compressed, and eigenvalues measure the amount of pull or compression along each axis.

### 4.4 SVD & Dimensionality Reduction

1. Create a random 4×3 matrix `X` (use `np.random.seed(42)`).
2. Compute its **SVD**: `U, S, Vt = np.linalg.svd(X, full_matrices=False)`.
3. **Reconstruct** `X` from the SVD and verify it matches the original.
4. Perform a **rank-1 approximation** using only the largest singular value/vector — call it `X_approx`.
5. Explain the relationship between SVD components and PCA in the markdown cell below.

````python
np.random.seed(42)
X = np.random.randn(4, 3)

# Economy SVD
U, S, Vt = np.linalg.svd(X, full_matrices=False)

# Reconstruct X
X_reconstructed = U @ np.diag(S) @ Vt

# Rank-1 approx
X_approx = S[0] * np.outer(U[:, 0], Vt[0, :])

print(f"U shape: {U.shape}, S shape: {S.shape}, Vt shape: {Vt.shape}")
print(f"Singular values: {S.round(4)}")
print(f"Reconstruction error: {np.linalg.norm(X - X_reconstructed):.2e}")
print(f"Rank-1 approximation:\n{X_approx.round(4)}")

assert np.allclose(X, X_reconstructed, atol=1e-10), "Reconstruction failed!"
print("4.4 passed ✓")
````

**SVD → PCA Connection (4.4):**

> In SVD, **X = U Σ Vᵀ**: the columns of **Vᵀ** (rows of Vt) are the **right singular vectors** — these are exactly the **principal component directions** (the eigenvectors of XᵀX). The columns of **U** are the **left singular vectors** — they are the principal component directions in sample space. The entries of **Σ** (singular values, S) relate to explained variance: the squared singular values, divided by (n − 1), are the eigenvalues of the covariance matrix. Therefore the **first row of Vt** points in the direction of **maximum variance** in the data. PCA is simply an economy SVD applied to a mean-centred data matrix — SVD is the numerically stable backbone that powers PCA.

### 5.2 Hypothesis Testing

### 5.3 Error Metrics

Given these predictions and actuals, **implement from scratch** (no sklearn):
- MAE, MSE, RMSE
- R² Score
- Adjusted R² (assume 2 features)

````python
y_true = np.array([3.0, 5.0, 2.5, 7.0, 4.5, 6.0, 1.5, 8.0])
y_pred = np.array([2.8, 5.2, 2.1, 7.5, 4.0, 6.3, 2.0, 7.8])
n, p = len(y_true), 2

# Residuals
residuals = y_true - y_pred

# MAE
mae      = np.mean(np.abs(residuals))

# MSE
mse      = np.mean(residuals ** 2)

# RMSE
rmse     = np.sqrt(mse)

# R squared
ss_res   = np.sum(residuals ** 2)
ss_tot   = np.sum((y_true - y_true.mean()) ** 2)
r2       = 1 - (ss_res / ss_tot)

# Adjusted R2
adj_r2   = 1 - (1 - r2) * (n - 1) / (n - p - 1)

print(f"MAE:        {mae:.4f}")
print(f"MSE:        {mse:.4f}")
print(f"RMSE:       {rmse:.4f}")
print(f"R²:         {r2:.4f}")
print(f"Adj. R²:    {adj_r2:.4f}")
````

### 5.4 Distribution Testing & Stationarity

1. Generate two samples: `s1` ~ Normal(0, 1), `s2` ~ Exponential(1), both n=200 (seed=0).
2. Run the **Kolmogorov-Smirnov test** (`stats.kstest`) comparing each against a standard normal CDF.
3. Generate a non-stationary time series (linear trend + noise). Run the **Augmented Dickey-Fuller test** (`adfuller` from `statsmodels`) and interpret the result.
4. Difference the series once and re-run ADF to confirm stationarity.

````python
from statsmodels.tsa.stattools import adfuller

np.random.seed(0)
s1 = np.random.normal(0, 1, 200)
s2 = np.random.exponential(1, 200)

# KS test
ks_s1 = stats.kstest(s1, 'norm', args=(0, 1))
ks_s2 = stats.kstest(s2, 'norm', args=(0, 1))
print(f"KS test s1 (normal):      stat={ks_s1.statistic:.4f}, p={ks_s1.pvalue:.4f}")
print(f"KS test s2 (exponential): stat={ks_s2.statistic:.4f}, p={ks_s2.pvalue:.4f}")

# ADF test
t = np.arange(200)
ts = 0.05 * t + np.random.normal(0, 1, 200)

adf_result = adfuller(ts)
print(f"\nADF statistic: {adf_result[0]:.4f}")
print(f"ADF p-value:   {adf_result[1]:.4f}")
print(f"Stationary?    {adf_result[1] < 0.05}")

# First difference
ts_diff = np.diff(ts)
adf_diff = adfuller(ts_diff)
print(f"\nAfter differencing — p-value: {adf_diff[1]:.4f}, Stationary? {adf_diff[1] < 0.05}")
````

````python
def compute_psi(expected, actual, bins=10):
    """
    PSI = sum((actual_pct - expected_pct) * ln(actual_pct / expected_pct))
    Add epsilon=1e-10 to avoid log(0).

    Steps:
      1. Build bin edges from the expected distribution.
      2. Count how many observations fall in each bin for both distributions.
      3. Convert counts to percentages (proportions).
      4. Apply the PSI formula element-wise and sum.
    """
    epsilon = 1e-10

    # Bin edges
    breakpoints = np.linspace(expected.min(), expected.max(), bins + 1)

    # Bin counts
    expected_counts, _ = np.histogram(expected, bins=breakpoints)
    actual_counts,   _ = np.histogram(actual,   bins=breakpoints)

    expected_pct = expected_counts / len(expected) + epsilon
    actual_pct   = actual_counts   / len(actual)   + epsilon

    # PSI formula
    psi_per_bin = (actual_pct - expected_pct) * np.log(actual_pct / expected_pct)
    return psi_per_bin.sum()


np.random.seed(1)
train_dist = np.random.normal(50, 10, 1000)    # training distribution
drift_dist = np.random.normal(65, 12, 1000)    # shifted production distribution

psi_value = compute_psi(train_dist, drift_dist)
print(f"PSI: {psi_value:.4f}")
print(f"Shift severity: {'Major' if psi_value > 0.2 else 'Minor' if psi_value > 0.1 else 'Stable'}")

# Plot distributions
fig, ax = plt.subplots()
ax.hist(train_dist, bins=30, density=True, alpha=0.5, color='steelblue',
        label='Training (μ=50, σ=10)')
ax.hist(drift_dist, bins=30, density=True, alpha=0.5, color='tomato',
        label='Production (μ=65, σ=12)')
ax.legend(); ax.set_title('Distribution Shift Visualization')
plt.tight_layout(); plt.show()
````

**Concept Drift vs Covariate Drift & PSI Thresholds (5.5):**

- **Concept drift:** The statistical relationship between features and the target changes over time — the model was trained on pattern X→Y, but in production the same X now maps to a different Y (e.g., customer behaviour changes post-pandemic, making old fraud rules obsolete).
- **Covariate drift:** The marginal distribution of the input features P(X) changes between training and production, even if the conditional P(Y|X) remains the same (e.g., a seasonal shift in user demographics).
- **PSI < 0.1:** No significant shift — the production distribution is stable and consistent with training data.
- **PSI 0.1–0.2:** Minor shift — flag for investigation; model performance may have degraded slightly.
- **PSI > 0.2:** Major shift — the input distribution has changed substantially; immediate model review and likely retraining is required.
- **Retraining trigger example:** Schedule an automatic retraining pipeline whenever the weekly PSI on any top-5 feature exceeds 0.2 for two consecutive monitoring windows.

### 6.1 Core Concepts

A bag contains: 4 red, 3 blue, 3 green marbles.

1. Define the sample space and compute P(red), P(blue), P(green).
2. Two marbles drawn **without replacement**. Compute:
   - P(first=red, second=blue) — **joint probability**
   - P(second=blue | first=red) — **conditional probability**
3. Are the two draws **independent**? Show mathematically.

````python
total = 10
red, blue, green = 4, 3, 3

# Base probabilities
p_red   = red   / total
p_blue  = blue  / total
p_green = green / total

# Joint probability
p_red_then_blue = (red / total) * (blue / (total - 1))

# Conditional probability
p_blue_given_red = blue / (total - 1)

# Independence check
independent = np.isclose(p_blue_given_red, p_blue)

print(f"P(red)={p_red:.3f}, P(blue)={p_blue:.3f}, P(green)={p_green:.3f}")
print(f"P(red,blue): {p_red_then_blue:.4f}")
print(f"P(blue|red): {p_blue_given_red:.4f}")
print(f"P(blue):     {p_blue:.4f}")
print(f"Independent? {independent}")
````

### 6.3 Bayes' Theorem

**Scenario — Spam filter:**
- P(Spam) = 0.30 (prior)
- P("free" | Spam) = 0.80 (likelihood)
- P("free" | Not Spam) = 0.05

1. Compute P(Spam | "free") using Bayes' theorem.
2. Implement a simple `naive_bayes_predict(prior_spam, p_word_given_spam, p_word_given_ham)` that returns the posterior P(Spam | word).
3. In the markdown cell, map each term — **prior, likelihood, evidence, posterior** — to the spam example.

````python
p_spam            = 0.30
p_free_given_spam = 0.80
p_free_given_ham  = 0.05
p_ham             = 1 - p_spam

# Total probability
p_free = (p_free_given_spam * p_spam) + (p_free_given_ham * p_ham)

# Bayes update
p_spam_given_free = (p_free_given_spam * p_spam) / p_free

print(f"P('free'): {p_free:.4f}")
print(f"P(Spam | 'free'): {p_spam_given_free:.4f}")


def naive_bayes_predict(prior_spam, p_word_given_spam, p_word_given_ham):
    """
    Compute the posterior probability that a message is spam given one word,
    using Bayes' theorem with the law of total probability for the evidence.

    Args:
        prior_spam        : P(Spam) — baseline spam rate
        p_word_given_spam : P(word | Spam) — likelihood of word appearing in spam
        p_word_given_ham  : P(word | Ham)  — likelihood of word appearing in legitimate mail

    Returns:
        float: posterior P(Spam | word)
    """
    prior_ham = 1 - prior_spam

    # Total probability
    evidence = (p_word_given_spam * prior_spam) + (p_word_given_ham * prior_ham)

    # Bayes update
    posterior = (p_word_given_spam * prior_spam) / evidence
    return posterior


pred = naive_bayes_predict(0.30, 0.80, 0.05)
assert abs(pred - p_spam_given_free) < 1e-9
print(f"naive_bayes_predict: {pred:.4f}")
print("6.3 passed ✓")
````

**Bayes Term Mapping (6.3):**

| Term | Formula | In this example |
|------|---------|------------------|
| Prior | P(Spam) | 0.30 — our belief that any incoming email is spam, before reading its content |
| Likelihood | P(word \| Spam) | 0.80 — how probable it is to see the word "free" if the email really is spam |
| Evidence | P(word) | 0.275 — the overall probability of the word "free" appearing in any email, regardless of class |
| Posterior | P(Spam \| word) | ≈ 0.873 — our updated belief that the email is spam *after* observing the word "free" |

### 6.4 Central Limit Theorem

1. Define a population following an **exponential distribution** (λ=1, size=100,000).
2. Repeatedly draw samples of size `n=30` and record their means — do this 5,000 times.
3. Plot the histogram of sample means.
4. Overlay the **theoretical normal distribution** predicted by the CLT: μ = population mean, σ = population_std / sqrt(n).
5. Run a KS test to confirm the distribution of sample means is approximately normal.

````python
np.random.seed(7)
population = np.random.exponential(scale=1.0, size=100_000)

n_samples   = 5000
sample_size = 30

# Sample means
sample_means = population[
    np.random.randint(0, len(population), size=(n_samples, sample_size))
].mean(axis=1)

pop_mean = population.mean()
pop_std  = population.std()

# CLT std
clt_std  = pop_std / np.sqrt(sample_size)

print(f"Population mean: {pop_mean:.4f}, Population std: {pop_std:.4f}")
print(f"Sample means mean: {sample_means.mean():.4f}, std: {sample_means.std():.4f}")
print(f"CLT predicted std: {clt_std:.4f}")

# Plot CLT
fig, ax = plt.subplots()
ax.hist(sample_means, bins=50, density=True,
        color='steelblue', alpha=0.6, edgecolor='white', label='Sample means')

x_range = np.linspace(sample_means.min(), sample_means.max(), 300)
ax.plot(x_range, stats.norm.pdf(x_range, pop_mean, clt_std),
        color='tomato', lw=2.5, label=f'CLT: N({pop_mean:.2f}, {clt_std:.2f})')

ax.set_title('CLT: Distribution of Sample Means')
ax.set_xlabel('Sample Mean'); ax.set_ylabel('Density')
ax.legend()
plt.tight_layout(); plt.show()

# KS test
ks_result = stats.kstest(sample_means, 'norm', args=(pop_mean, clt_std))
print(f"KS test p-value: {ks_result.pvalue:.4f} → Approximately normal? {ks_result.pvalue > 0.05}")
````

**CLT Reflection (6.4):**

> The Central Limit Theorem guarantees that, regardless of the shape of the underlying population distribution, the distribution of sample means will converge to a normal distribution as sample size grows. This is foundational to statistics and ML because it justifies using z-tests and t-tests (which assume normality) even when the raw data is skewed — we are testing the *mean*, not the raw values. It also underpins confidence intervals and standard errors in almost every classical estimator, meaning practitioners can apply normal-distribution machinery broadly without needing the raw data to be Gaussian.
